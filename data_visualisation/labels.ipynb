# Data Visualization
## Segment 4 - Creating labels and annotations

```python
import pandas as pd
import seaborn as sb

from matplotlib import pyplot
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
sb.set_style('whitegrid')
```

### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

pyplot.bar(x,y)

# Set the x and y axis labels using xlabel and ylabel
pyplot.xlabel('This is the x axis')
pyplot.ylabel('This is the y axis')
```

```python
# Pie charts can have labels assigned to them by setting the labels arg equal to a list of labels

z = [1,2,3,4,0.5]
vehicle_types = ['bicycle', 'motorbike', 'car', 'van', 'stroller']
pyplot.pie(z, labels = vehicle_types)
pyplot.show
```

#### The object-oriented method

```python
#columns can be accessed as series objects using dot notation as well a square brackets
cars = pd.read_csv('./mtcars.csv')
mpg = cars.mpg

figure  = pyplot.figure()
axes = figure.add_axes([.1,.1,1,1])
mpg.plot()

# The axis for the figure can be manipulated in the following way
axes.set_xticks(range(32))
axes.set_xticklabels(cars.car_names, rotation = 80, fontsize = 'medium')
axes.set_xlabel('Cars')
axes.set_ylabel('mpg')
```

### Adding a legend to your plot
#### The functional method

```python
# The legend method can be called from the pyplot module
# pyplot.pie(z)
pyplot.legend(vehicle_types, loc = 'best')
```

#### The object-oriented method

```python
# This first section is just to replecate the above line chart
figure  = pyplot.figure()
axes = figure.add_axes([.1,.1,1,1])
mpg.plot()
axes.set_xticks(range(32))
axes.set_xticklabels(cars.car_names, rotation = 80, fontsize = 'medium')
axes.set_xlabel('Cars')
axes.set_ylabel('mpg')

# The legend can also be set on the axis module
axes.legend(loc = 'best')

```

### Annotating your plot

```python
figure  = pyplot.figure()
axes = figure.add_axes([.1,.1,1,1])
mpg.plot()

# Set the graph title
axes.set_title('Graph to show ....')

# Set the y axis label
axes.set_ylabel('miles/gal')

# Set the y axis range
axes.set_ylim(0,45)

# Add an annotation
# Frist arg is the text to show in the annotation
# 'xy'  = corodiantes of point to annotate
# 'xytext' is location you want text to be in
# 'arrowprops' is the properties of the arrow generated

axes.annotate('Toyota Carolla', 
              xy = (19,33.9), 
              xytext = (21, 35),
              arrowprops = dict(facecolor = 'powderblue', shrink = 0.05))
```
