## Segment 5 - Creating visualizations from time series data

```python
import pandas as pd
import seaborn as sb

from matplotlib import pyplot
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = [10,8]
sb.set_style('whitegrid')
```

### The simplest time series plot

```python
# We can index a data frame using the index_col arg.
# We also have to tell pandas to parse the dates if we want it to.

data = pd.read_csv('./Superstore.csv', index_col = 'Order Date', parse_dates = True)
data.head()
```

```python
data['Order Quantity'].plot()
```

```python
sample = data.sample(n=100, random_state=25, axis=0)
pyplot.xlabel('Order Date')
pyplot.ylabel('Order Quantity')
pyplot.title('Sales')
sample['Order Quantity'].plot()
```
