#### Note - Steps to create an OOP plot

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from matplotlib import pyplot
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

### Defining axes, ticks, and grids

```python
x = range(1,10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

# Create the blank figure
figure = pyplot.figure()

# Specifiy the axis
# The list arg takes the format of [left, bottom, wigth, height]
axes = figure.add_axes([0.1, 0.1, 1, 1])

axes.plot(x,y)
```

```python
figure = pyplot.figure()
axes = figure.add_axes([0.1, 0.1, 1, 1])

# Set the maximum values for x and y axis
axes.set_xlim([0,10])
axes.set_ylim([0,6])

# Set where the x and y ticks should be placed 
axes.set_xticks([0,1,2,3])
axes.set_yticks([0,1,3,4,5,6])

axes.plot(x,y)
```

```python
figure = pyplot.figure()
axes = figure.add_axes([0.1, 0.1, 1, 1])
axes.set_xlim([0,10])
axes.set_ylim([0,6])

# Adds a grid to the background of the graph
axes.grid()

axes.plot(x,y)
```

### Adding labels and titles to figures

```python
figure = pyplot.figure()
axes = figure.add_axes([0.1, 0.1, 1, 1])
axes.set_xlim([0,10])
axes.set_ylim([0,6])

# Add x and y labels to axes
axes.set_xlabel('X axis / units')
axes.set_ylabel('y axis / units')

# Add title to figure
axes.set_title('Plot title')

axes.plot(x,y)
```
